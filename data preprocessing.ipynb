```python

# Impute missing 'age' with the mean of the column
df['age'] = df['age'].fillna(df['age'].mean())

# Impute missing 'gender' with the mode (most frequent value) of the column
df['gender'] = df['gender'].fillna(df['gender'].mode()[0])

# Save the cleaned DataFrame
df.to_csv("travel_recommendations_cleaned.csv", index=False)
```

```python
import pandas as pd

# Load your cleaned dataset
df = pd.read_csv("C:/Users/ammar/Downloads/travel_recommendations_cleaned.csv")  # <-- update if filename differs

# -----------------------------
# Step 1: Label Encoding for 'budget'
# -----------------------------
# 'budget' has an inherent order: Low < Medium < High, so Label Encoding is appropriate
budget_mapping = {'Low': 0, 'Medium': 1, 'High': 2}
df['budget'] = df['budget'].map(budget_mapping)

# -----------------------------
# Step 2: One-Hot Encoding for Nominal Categorical Columns
# -----------------------------
# These columns have categories with no natural order, so we use One-Hot Encoding
one_hot_columns = ['gender', 'hobby', 'travel_style', 'destination_country', 'climate']
df = pd.get_dummies(df, columns=one_hot_columns)

# -----------------------------
# Step 3: Multi-Label One-Hot Encoding for 'activities'
# -----------------------------
# 'activities' contains comma-separated values like "A, B, C", so we need to create binary columns for each activity

# Step 3.1: Split activities into lists
df['activities_split'] = df['activities'].str.split(', ')

# Step 3.2: Get all unique activities
all_activities = set([activity for sublist in df['activities_split'] for activity in sublist])

# Step 3.3: Create binary columns for each activity
for activity in all_activities:
    df[f'activity_{activity}'] = df['activities_split'].apply(lambda x: int(activity in x))

# Step 3.4: Drop the original 'activities' columns
df.drop(['activities', 'activities_split'], axis=1, inplace=True)

# -----------------------------
# Step 4: - Drop 'user_id' (not useful for content-based filtering)
# -----------------------------
df.drop('user_id', axis=1, inplace=True)

# -----------------------------
# Final Step: Save the fully encoded dataset
# -----------------------------
df.to_csv("C:/Users/ammar/Downloads/final_encoded_dataset.csv", index=False)

print("✅ Dataset encoded and saved successfully!")
```
